```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
l1_means = []
l1_std_dev = []
l2_means = []
l2_std_dev = []
x_axis = []
for i in range(11):
  data = np.random.rand(100, 2**i)
  l1_distance_list = []
  l2_distance_list = []
  for j in range(100):
    for k in range(j+1, 100):
      l1_dist = 0
      l2_dist = 0
      for pos in range(2 ** i):
        l1_dist += abs(data[j][pos] - data[k][pos])
        l2_dist += (data[j][pos] - data[k][pos]) * (data[j][pos] - data[k][pos])
      l1_distance_list.append(l1_dist)
      l2_distance_list.append(l2_dist)
  l1_means.append(np.mean(l1_distance_list))
  l1_std_dev.append(np.std(l1_distance_list))
  l2_means.append(np.mean(l2_distance_list))
  l2_std_dev.append(np.std(l2_distance_list))
  x_axis.append(2 ** i)
print(l1_means)
print(l1_std_dev)
print(l2_means)
print(l2_std_dev)

  
```

```python
plt.plot(x_axis, l1_means, label="l1 Average Distance")
plt.xlabel("dimension")
# plt.legend()
plt.show()
```

```python
plt.plot(x_axis, l1_std_dev, label="l1 Standard Deviation")
plt.xlabel("dimension")
plt.legend()
plt.show()
```

```python
plt.plot(x_axis, l2_means, label="l2 Average Distance")
plt.xlabel("dimension")
# plt.legend()
plt.show()
```

```python
plt.plot(x_axis, l2_std_dev, label="l2 Standard Deviation")
plt.xlabel("dimension")
plt.legend()
plt.show()
```
